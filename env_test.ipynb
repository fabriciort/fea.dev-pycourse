```python
import matplotlib.pyplot as plt
import numpy as np

def fibonacci(n):
  """
  Calcula a sequência de Fibonacci até o termo n.

  Args:
    n: O termo n da sequência de Fibonacci.

  Returns:
    Uma lista com os primeiros n termos da sequência de Fibonacci.
  """

  fibo = [0, 1]
  for i in range(2, n + 1):
    fibo.append(fibo[i - 1] + fibo[i - 2])
  return fibo

def plot_fibonacci(n):
  """
  Plota uma curva Fibonacci até o termo n.

  Args:
    n: O termo n da sequência de Fibonacci.
  """

  x = np.linspace(0, n, n + 1)
  y = fibonacci(n)
  plt.plot(x, y)
  plt.xlabel('Termo')
  plt.ylabel('Número de Fibonacci')
  plt.show()


if __name__ == '__main__':
  # Exemplo de uso
  plot_fibonacci(75)
```
